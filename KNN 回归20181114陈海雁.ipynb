```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Iris.csv',header=0)
data.drop(labels=['Id','Species'],axis=1,inplace=True)
data.drop_duplicates(inplace=True)
#data.head()

```

```python
def predict(train_X,train_y,test_X,k):
    train_X=np.asarray(train_X)
    train_y=np.asarray(train_y)
    test_X=np.asarray(test_X)
    result=[]
    for x in test_X:
        dis=np.sqrt(np.sum((x-train_X)**2,axis=1))
        index=np.argsort(dis)
        index=index[:k]
        jieguo=np.mean(train_y[index])
        result.append(jieguo)
    return np.asarray(result)


#加入权重
def predict2(train_X,train_y,test_X,k):
    train_X=np.asarray(train_X)
    train_y=np.asarray(train_y)
    test_X=np.asarray(test_X)
    result=[]
    for x in test_X:
        dis=np.sqrt(np.sum((x-train_X)**2,axis=1))
        index=np.argsort(dis)
        index=index[:k]
        
        weight=1/dis[index]/np.sum(1/dis[index])
        result.append(np.sum(train_y[index]*weight))
    return np.asarray(result)
```

```python
t=data.sample(n=len(data),random_state=0)
train_X=t.iloc[:100,:-1]
train_y=t.iloc[:100,-1]
test_X=t.iloc[100:,:-1]
test_y=t.iloc[100:,-1]
test_y=np.asarray(test_y)

result=predict(train_X,train_y,test_X,3)
error=(result-test_y)
```

```python
#加入权重
result=predict2(train_X,train_y,test_X,3)
plt.figure(figsize=(12,8))
plt.plot(test_y,c='b',marker='o',label='test')
plt.plot(result,c='r',marker='o',label='result')
plt.show()
```
